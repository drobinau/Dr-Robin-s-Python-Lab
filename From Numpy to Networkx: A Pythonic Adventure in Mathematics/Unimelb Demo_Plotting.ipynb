### Example 1: Plotting a logistic sequence 
$$x_{n+1} = a x_n(1-x_n)$$

```python
# Logistic equation
import matplotlib.pyplot as plt
alist = [0.7]  # Initial value for x_0
a = 2  
for i in range(10):
    t = a * alist[-1] * (1 - alist[-1])  # Calculate next term
    alist.append(t)  # Append the new value to the list
print(alist)  
 
# Plot the sequence
plt.plot(alist, '.', label='Logistic Sequence')
plt.xlabel('n')  
plt.ylabel('x_n')  
plt.title('Logistic Map')  
plt.legend()  
plt.show()  
```

### Example 2: Customising plots 
$$ y = sin(x)\\
y = cos(x)$$

```python
import matplotlib.pyplot as plt
import numpy as np
 
# Create an array of x values from 0 to 2Ï€
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
 
plt.plot(x, y1, label='Sine Function', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label='Cosine Function', color='red', linestyle='--', linewidth=2)
 
plt.title('Sine and Cosine Functions')
plt.xlabel('x')
plt.ylabel('y') 
plt.legend()
plt.grid(True)

plt.show()
```

### Example 3: Subplots


```python
import matplotlib.pyplot as plt
import numpy as np
 
# Create a range of x values
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
 
# Create a 1x2 grid of subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
 
# Plot the sine function in the first subplot
axes[0].plot(x, y1, label='sin(x)', color='blue')
axes[0].set_title('Sine Function')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
 
# Plot the cosine function in the second subplot
axes[1].plot(x, y2, label='cos(x)', color='green')
axes[1].set_title('Cosine Function')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
 
# Display the plots
plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
