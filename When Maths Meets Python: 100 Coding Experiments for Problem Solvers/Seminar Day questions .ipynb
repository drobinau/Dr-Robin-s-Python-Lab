#### Two ordinary dice are tossed and the difference between the numbers appearing uppermost on the dice is recorded. What difference is most likely to occur?

```python
import random
import matplotlib.pyplot as plt

n = 10000  # Number of simulations
count = [0] * 6  # List to count occurrences of each difference

for i in range(n):
    dice_a = random.randint(1, 6)  # Roll the first die
    dice_b = random.randint(1, 6)  # Roll the second die
    diff_c = abs(dice_a - dice_b)  # Calculate the absolute difference

    count[diff_c] += 1  # Increment the count for the corresponding difference

# Print the number of occurrences for each difference
print('The number of each difference is:', count)

# Determine and print the most likely difference
most_likely_difference = count.index(max(count))
print('The most likely difference is:', most_likely_difference)

# Plotting the histogram
differences = list(range(6))  # 0 to 5
plt.bar(differences, count, color='skyblue', edgecolor='black')
plt.title('Distribution of Absolute Differences (Two Dice Rolls)')
plt.xlabel('Absolute Difference')
plt.ylabel('Frequency')
plt.xticks(differences)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Two ordinary dice are tossed and the difference between the numbers appearing uppermost on the dice is recorded. What sum is most likely to occur?

```python
import random
import matplotlib.pyplot as plt

n = 10000  # Number of simulations
count = [0] * 13  # Index 0â€“12 (we'll ignore 0 and 1; sums go from 2 to 12)

for i in range(n):
    dice_a = random.randint(1, 6)  # Roll the first die
    dice_b = random.randint(1, 6)  # Roll the second die
    sum_s = dice_a + dice_b  # Calculate the sum

    count[sum_s] += 1  # Increment the count for the corresponding sum

# Print the number of occurrences for each sum from 2 to 12
print('The number of each sum (2 to 12) is:', count[2:])

# Determine and print the most likely sum
most_likely_sum = count.index(max(count))
print('The most likely sum is:', most_likely_sum)

# Plotting the histogram
sums = list(range(2, 13))  # Possible sums from 2 to 12
frequencies = count[2:]  # Ignore index 0 and 1

plt.bar(sums, frequencies, color='lightgreen', edgecolor='black')
plt.title('Distribution of Sums (Two Dice Rolls)')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Frequency')
plt.xticks(sums)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### What is the most likely number of rolls until the first repeat of the first face up number? 

```python
import random
import matplotlib.pyplot as plt

n = 10000  # Number of trials
roll_counts = []  # To store number of rolls per trial

for _ in range(n):
    first_roll = random.randint(1, 6)
    count = 1

    while True:
        roll = random.randint(1, 6)
        count += 1
        if roll == first_roll:
            roll_counts.append(count)
            break

# Calculate average
average_rolls = sum(roll_counts) / n
print("The average number of rolls until repeat is:", average_rolls)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(roll_counts, bins=range(min(roll_counts), max(roll_counts) + 2), edgecolor='black', align='left')
plt.axvline(average_rolls, color='red', linestyle='dashed', linewidth=1.5, label=f'Average = {average_rolls:.2f}')
plt.title("Distribution of Rolls Until Repeat of First Face")
plt.xlabel("Number of Rolls")
plt.ylabel("Frequency")
plt.xticks(range(min(roll_counts), max(roll_counts) + 1))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### I first roll a standard die, then I toss a pair of fair coins as many times as the die indicates. If the pair of coins show two heads in any of the tosses, then you win the game. Otherwise, I win the game. Should you play this game?

```python
import random
 
n = 1000  # Number of trials
count = 0
 
for i in range(n):
    # Roll a six-sided die to determine the number of tosses
    a = random.randint(1, 6)
    # Simulate the coin tosses
    b = [(random.randint(0, 1), random.randint(0, 1)) for _ in range(a)]
    
    # Check if two heads appeared in any of the tosses
    if (1, 1) in b:
        count += 1
 
# Calculate probability
print('The probability that two heads appear in the outcome is: {}%, {}/{}'.format(count / n * 100, count, n))
```
